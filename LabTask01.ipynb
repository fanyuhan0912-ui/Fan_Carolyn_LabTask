```python
df["Fare_Euro"] = df["Fare"] * 0.92

df.head()
```

```python
df["Name_Length"] = df["Name"].apply(len)

df.head()
```

```python
df["Fare_Age_Diff"] = df["Fare"] - df["Age"]

df.head()
```

```python
df.round(1)
```

Explanation:

apply(len) counts the number of characters in each passenger name.

round(1) rounds all numerical values to one decimal place.

```python
df["Sex_Code"] = df["Sex"].map({
    "male": 0,
    "female": 1
})

df.head()
```

```python
def age_group(age):
    if pd.isna(age):
        return "unknown"
    elif age < 18:
        return "child"
    elif age < 60:
        return "adult"
    else:
        return "senior"
```

```python
df["Age_Group"] = df["Age"].apply(age_group)

df.head()
```

Explanation:

map() converts categorical values into numeric codes.

apply() is used to apply the custom age_group function to every Age value.

```python
import csv
import time
```

```python
# Pandas version
start = time.time()

pandas_single_row = df.iloc[0]

end = time.time()
pandas_time = end - start

print(pandas_single_row)
print("Pandas time:", pandas_time)
```

```python
# Pure Python version
start = time.time()

data_list = df.to_dict("records")
python_single_row = data_list[0]

end = time.time()
python_time = end - start

print(python_single_row)
print("Pure Python time:", python_time)
```

```python
# Pandas version
start = time.time()

pandas_age_above_60 = df[df["Age"] > 60]

end = time.time()
pandas_time = end - start

print(pandas_age_above_60)
print("Pandas time:", pandas_time)
```

```python
# Pure Python version
start = time.time()

data_list = df.to_dict("records")

python_age_above_60 = []

for row in data_list:
    if row["Age"] > 60:
        python_age_above_60.append(row)

end = time.time()
python_time = end - start

print(python_age_above_60)
print("Pure Python time:", python_time)
```

```python
# Pandas version
start = time.time()

df["Fare_Euro_Pandas"] = df["Fare"] * 0.92

end = time.time()
pandas_time = end - start

print(df[["Fare", "Fare_Euro_Pandas"]].head())
print("Pandas time:", pandas_time)
```

```python
# Pure Python version
start = time.time()

data_list = df.to_dict("records")

for row in data_list:
    row["Fare_Euro_Python"] = row["Fare"] * 0.92

end = time.time()
python_time = end - start

print(data_list[:5])
print("Pure Python time:", python_time)
```
